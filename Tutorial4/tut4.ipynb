```python
# Import libraries
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)

%matplotlib inline
```

```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
```

```python
URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB3.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path): # this extraction only happens if folder is not present
    download_and_unzip(URL, asset_zip_path)   
```

```python
# Read in an image
image = cv2.imread("Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)

# Display the original image
plt.subplot(2,1,1)
plt.imshow(image[:, :, ::-1])
plt.subplot(2,1,2)
plt.imshow(image)
```

```python
imageLine = image.copy()

# The line starts from (200,100) and ends at (400,100)
# The color of the line is YELLOW (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

#cv2.line(imageLine, (200, 100), (400, 100), (0, 255, 255), thickness=1, lineType=cv2.FILLED)
cv2.line(imageLine, (200, 100), (400, 100), (0, 255, 255), thickness=2, lineType=cv2.LINE_4)
cv2.line(imageLine, (200, 200), (400, 100), (0, 255, 255), thickness=2, lineType=cv2.LINE_8)
cv2.line(imageLine, (200, 300), (400, 100), (0, 255, 255), thickness=2, lineType=cv2.LINE_AA)


# Display the image
plt.imshow(imageLine[:,:,::-1])
```

```python
# Draw a rectangle (thickness is a positive integer)
imageRectangle = image.copy()
# only diagonal points are specified
cv2.rectangle(imageRectangle, (500, 100), (700, 600), (255, 0, 255), thickness=5, lineType=cv2.LINE_8)

# Display the image
plt.imshow(imageRectangle[:, :, ::-1])
```

```python
# Draw a circle
imageCircle = image.copy()
#center point and radious is mentioned
cv2.circle(imageCircle, (900,500), 200, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA)

# Display the image
plt.imshow(imageCircle[:,:,::-1])
```

```python
imageText = image.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
#fontScale = 2.3
#fontFace = cv2.FONT_HERSHEY_PLAIN
#ontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (200, 700), cv2.FONT_HERSHEY_PLAIN, 2.3, (0,255,0), fontThickness, cv2.LINE_AA);

# Display the image
plt.imshow(imageText[:, :, ::-1])
```

```python
import cv2
import numpy as np

# Create a blank image
image = np.zeros((500, 500, 3), dtype=np.uint8)

# Use positive font scale (normal behavior)
cv2.putText(image, 'Positive Font Scale', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)

# Use negative font scale (text won't be rendered)
cv2.putText(image, 'Negative Font Scale', (50, 200), cv2.FONT_HERSHEY_SIMPLEX, -1.0, (255, 255, 255), 2)

# Display the image
cv2.imshow('Text Example', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
